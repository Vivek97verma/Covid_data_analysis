```python
# load the necessary library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# load the datset
df=pd.read_csv('covid.csv')
df.head()
```

```python
#fetch the column
df.columns
```

```python
# check the data type of features
df.dtypes
```

```python
# change the data type of "date" to datetime data type
df['date']=pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
# check the shape of data frame
df.shape
```

```python
df.dtypes.value_counts()
```

```python
# Extracting quarter column from date column
df['qtr']=df['date'].dt.quarter
df.head()
```

```python
df['qtr'].value_counts()
```

# KDE plot

```python
sns.kdeplot(data=df,x='cases')
plt.title('kde plot for covid cases')
plt.show()
```

```python
# Q- plot kde plot for the cases for the month of march and april.
df_mar_apr=df[df['month'].isin([3,4])]
df_mar_apr['month'].value_counts()
```

```python
df_mar_apr.shape
```

```python
sns.kdeplot(data=df_mar_apr,x='cases',hue='month',palette='bright')
plt.show()
```

# Histplot

```python
# plot hist plot for cases for the countries- india,belgium and russia for the months including mar and apr.
df_countries=df[df['countries'].isin(['India','Belgium','Russia'])]
df_countries
```

```python
df_3c_mar_apr=df_countries[df_countries['month'].isin([3,4])]
df_3c_mar_apr
```

```python
sns.histplot(data=df_3c_mar_apr,x='cases')
plt.title('histplot for the cases in march and april')
plt.show()
```

```python
# plot hist plot for cases for the countries- india,belgium and russia for the months including mar and apr. where month is set as hue parameter
sns.histplot(data=df_3c_mar_apr,x='cases',hue='month',palette='bright')
plt.title('histplot for the month of march and april for india,belgium and russia') 
plt.show()
```

```python
# plot histplot for deaths for the countries-india,belgium and russia where month is march and april
sns.histplot(data=df_3c_mar_apr,x='deaths')
plt.show()
```

```python
# set month as hue parameter
sns.histplot(data=df_3c_mar_apr,x='deaths',hue='month',palette='bright')
plt.title("Histplot for deaths in march and april for india, belgium and russia")
plt.show()
```

# Heatmap

###syntax= sns.heatmap(data,annot,cmap)

```python
# plot the correlation matrix on heatmap,excluding the column year
df.columns
```

```python
df1=df.iloc[:,[0,1,2,4,5,10]]
df1.columns
```

```python
corr=df1.corr()
corr
```

```python
sns.heatmap(data=corr,annot=True)
plt.show()
```

```python
sns.heatmap(data=corr,annot=True,cmap='RdBu')
plt.show()
```

```python
corr_cd=df.iloc[:,[4,5]].corr()
corr_cd
```

```python
sns.heatmap(data=corr_cd,annot=True,cmap='coolwarm')
plt.show()
```

# Joint plot

### it is used to plot the value of two variables with bivariate and univaritae graphs. it is a combination of scatter plot and histogram

```python
# plot a joint plot between cases and deaths for the countries- India ,Belgium and Russia for the months including mar and apr
sns.jointplot(data=df_3c_mar_apr,x='cases',y='deaths')
plt.show()
```

```python
sns.jointplot(data=df_3c_mar_apr,x='cases',y='deaths',hue='month',palette='bright')
plt.show()
```

```python
sns.jointplot(data=df_3c_mar_apr,x='cases',y='deaths',hue='month',palette='bright',kind='kde')
plt.show()
```

```python
sns.jointplot(data=df_3c_mar_apr,x='cases',y='deaths',hue='month',palette='bright',kind='hist')
plt.show()
```

```python
sns.jointplot(data=df_3c_mar_apr,x='cases',y='deaths',palette='bright',kind='reg')
plt.show()
```

```python
sns.jointplot(data=df_3c_mar_apr,x='cases',y='deaths',palette='bright',kind='hex',color='maroon')
plt.show()
```

```python
sns.jointplot(data=df_3c_mar_apr,x='cases',y='deaths',palette='bright',kind='resid')
plt.show()
```
